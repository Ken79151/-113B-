取得 Name 和 Salary 二欄資料

```python

import pandas as pd


data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)

df

```

請問公司人員的平均薪水、及最高、最低薪水、中位數？

```python
print(df['Salary'].mean(),df['Salary'].max(),df['Salary'].min(),df['Salary'].median())
```

```python
df['Company'] = 'TechCorp'
df
```

```python
df['Annual Salary'] = df['Salary']* 13.5
df
```

```python
#新增高薪欄位 (薪資高於65000 標記yes 否則no )
df['High Salary'] = df['Salary'].apply(lambda x: 'Yes' if x > 65000 else 'No')
df
```

```python
df.drop(columns=['High Salary'],inplace=True)
df

#另一種方法
df= df.drop(columns=['High Salary'])
```

```python
# 使用串列方式新增一筆資料（按照 DataFrame 欄位順序）
df.loc[len(df)] = [111, 'Kevin', 28, 'Austin', 67000, 'IT','TechCorp',360000]
df
```

```python
new_data = [
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
new_df= pd.DataFrame(new_data)
new_df
```

```python
df = pd.concat([df,new_df],ignore_index=True)
df
```

```python
# 刪除索引為 2 的資料
df= df.drop(index= 2, inplace=True)
df
```

```python
condition = df['Salary'] < 75000
df = df[condition]
df
```

```python
# 計算每一欄有多少個 NaN
df.isnull().sum()
```

```python
# 找出具有空值的列
df[['ID','Name']].isnull().any(axis=1)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[4,'Salary'] = pd.NA
df
```

```python
# 用平均數填充 Salary 欄位
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df
```

```python
# 替換 'IT' 部門為 'Information Technology'
df['Department'] = df['Department'].replace('IT', 'Information Technology')
df
```

```python
# 如果 City 是 'New York'，則改為 'NYC'
df['City'] = df['City'].apply(lambda x: 'NYC' if x == 'New York' else x)
```

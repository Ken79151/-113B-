#背景

#目的

#數據來源
本研究使用 TensorFlow Keras 提供的 Boston Housing 數據集，包含多項波士頓城鎮特徵與對應房價中位數。

#分析工具

Python（數據處理與分析）

Pandas（資料整理與統計分析）

Matplotlib / Seaborn（數據視覺化）

#數據處理

載入資料並檢視基本統計與資料完整性。

進行散點圖分析房價與關鍵變數的關係。

計算相關係數，量化變數間關聯強度。

透過條件篩選，找出犯罪率低、距離就業中心近且房間數多的理想居住地區。



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import boston_housing

# 載入 Boston Housing 數據集
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
           'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
df = pd.DataFrame(x_train, columns=columns)
df['MEDV'] = y_train  # 房價中位數 (千美元)

# 顯示數據前五筆
print(df.head())

# 顯示數據基本資訊
df.info()

#距離波士頓五個就業中心的距離(DIS)vs房價中位數(MEDV)
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='DIS', y='MEDV')
plt.title('Distance to Employment Centers vs Median House Price')
plt.xlabel('DIS (Distance to Employment Centers)')
plt.ylabel('MEDV (Median House Price in $1000s)')
plt.show()

#城鎮犯罪率(CRIM)vs房價中位數(MEDV)
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='CRIM', y='MEDV')
plt.title('Crime Rate vs Median House Price')
plt.xlabel('CRIM (Per Capita Crime Rate)')
plt.ylabel('MEDV (Median House Price in $1000s)')
plt.show()

#平均房間數(RM)vs房價中位數
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='RM', y='MEDV')
plt.title('Average Number of Rooms vs Median House Price')
plt.xlabel('RM (Average Number of Rooms)')
plt.ylabel('MEDV (Median House Price in $1000s)')
plt.show()

#計算相關係數
corrs = df[['DIS', 'CRIM', 'RM', 'MEDV']].corr()['MEDV']
print("Correlation coefficients with MEDV:")
print(corrs)

#篩選適合居住地區
ideal_df = df[(df['CRIM'] < 1) & (df['DIS'] < 5) & (df['RM'] > 6) & (df['MEDV'] > 25)]
print(f"\nNumber of recommended residential areas: {len(ideal_df)}")
print(ideal_df.head())

#適合居住地區視覺化
plt.figure(figsize=(8,5))
sns.scatterplot(data=ideal_df, x='CRIM', y='MEDV', hue='RM', palette='viridis', size='DIS', sizes=(20, 200))
plt.title("Recommended Residential Areas")
plt.xlabel("Crime Rate (CRIM)")
plt.ylabel("Median House Price (MEDV in $1000s)")
plt.legend(title='Average Rooms (RM)')
plt.show()

#分析總結
summary = """
分析結果：
- 房價中位數 (MEDV) 隨著到就業中心距離 (DIS) 增加而下降，表示越靠近就業中心，房價越高。
- 城鎮犯罪率 (CRIM) 與房價呈負相關，犯罪率越低的區域，房價越高。
- 平均房間數 (RM) 與房價呈現強正相關，房間數越多，房價越高。
- 綜合分析後，犯罪率低於1、距離就業中心少於5、平均房間數大於6，且房價中位數超過25（千美元）為適合居住的地區篩選標準。

結論：
本次分析指出，安全性高、鄰近就業中心且居住空間寬敞的地區更具吸引力且房價較高，這些地區適合居住與長期投資。

建議：
- 買房者應優先考慮上述條件篩選區域，以獲得較佳生活品質及房產保值。
- 政府應加強低犯罪率區域的公共設施與交通建設，提高生活便利性。
- 未來研究可納入更多社會經濟指標，深入分析影響房價與居住品質的其他因素。
"""

print(summary)
```
